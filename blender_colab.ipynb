```python
from re import DEBUG
import subprocess
import json
from os.path import join, realpath, split
import argparse
import logging


def render(blend_file, blender_path='blender'):
    """Open the blend file and generate the samples specified in the file"""

    # command = './blender/blender --background ' + config['blender_scene'] +' --python generator.py -- -c '+ config_path
    script_path = '/content/auxfunc.py'
    command = f'{blender_path} {blend_file} --background --python {script_path}'

    p = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    for line in p.stdout:
        print(line.decode('utf-8').rstrip())
    retval = p.wait()
```

```python
blender_path = f'{blender_version}/blender'
render(path_to_blend, blender_path)
```
